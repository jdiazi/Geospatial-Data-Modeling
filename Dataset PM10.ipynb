## Class 6: URL importing- PM10 dataset example

The file title PM10 has information on pollutants measured in the air through 43 monitoring stations, located in 3 provinces of the Basque Country, Spain.

Find the dataset here:  https://wp.me/aa4oIl-2eP

```python
import pandas as pd
```

```python
df = pd.read_excel('https://wp.me/aa4oIl-2eP')
df.head()
```

```python
df.count()
```

### Group dataset by Province

```python
df2=df.groupby(['Province'])
df2.count()
```

### Creating new df based on each province
#### Method 1:

```python
df_Araba = df[df['Province']=='Araba/Álava']
df_Araba.head()
```

```python
df_Biskaia = df[df['Province']=='Bizkaia']
df_Biskaia.head()
```

```python
df_Gipuzkoa = df[df['Province']=='Gipuzkoa']
df_Gipuzkoa.head()
```

#### Method 2:

```python
df_Araba2=df2.get_group('Araba/Álava')
df_Araba2.head()
```

```python
df_Bizkaia2=df2.get_group('Bizkaia')
df_Bizkaia2.head()
```

```python
df_Gipuzkoa2=df2.get_group('Gipuzkoa')
df_Gipuzkoa2.head()
```

#### Method 3:

```python
df_sub = df.loc[:, ['NO (µg/m3)', 'NO2 (µg/m3)', 'NOX (µg/m3)', 'PM10 (µg/m3)', 'Province']]
df_sub.head()
```

#### Method 4:
